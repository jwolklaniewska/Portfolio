# MIN-MAX METHOD
Metoda polegająca na sporządzeniu wykresu zależności maksimum od minimum, aby zobrazować zależność ekstremum względem siebie

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
from statsmodels.distributions.empirical_distribution import ECDF
import math
from scipy.stats import genextreme as gev
import seaborn as sns
from fitter import Fitter
```

```python
PD_eyesOpen = np.load('PD_eyesOpen.npy', allow_pickle=True)
PD_eyesClosed = np.load('PD_eyesClosed.npy', allow_pickle=True)

HC_eyesOpen = np.load('HC_eyesOpen.npy', allow_pickle=True)
HC_eyesClosed = np.load('HC_eyesClosed.npy', allow_pickle=True)
```

```python
window = 100
```

# OTWARTE OCZY

```python
for i in range(len(PD_eyesOpen)):
    data = PD_eyesOpen[i]
    l = len(data)
    maximums = np.zeros(len(data)//window+1)
    minimums = np.zeros(len(data)//window+1)

    k = 0
    while k*window < l:
        d = data[k*window:(k+1)*window]
        maximums[k] = max(d)
        minimums[k] = min(d)
        # print(f"k={k} Done")
        k += 1

    x = np.linspace(min(minimums), max(minimums), 1000)
    

    plt.scatter(minimums, maximums)
    plt.xlabel('minimum')
    plt.ylabel('maximum')
    plt.plot(x,x)
plt.show()

for i in range(len(PD_eyesOpen)):
    data = HC_eyesOpen[i]
    l = len(data)
    maximums = np.zeros(len(data)//window+1)
    minimums = np.zeros(len(data)//window+1)

    k = 0
    while k*window < l:
        d = data[k*window:(k+1)*window]
        maximums[k] = max(d)
        minimums[k] = min(d)
        # print(f"k={k} Done")
        k += 1

    x = np.linspace(min(minimums), max(minimums), 1000)
    

    plt.scatter(minimums, maximums)
    plt.xlabel('minimum')
    plt.ylabel('maximum')
    plt.plot(x,x)
plt.show()
```

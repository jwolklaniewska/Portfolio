# OGONY ROZKŁADU W SKALI PODWÓJNIE LOGARYTMICZNEJ

```python
n1 = 5
n2 = 1

s = PD_eyesOpen[n1]
s = s[(s > 0)]
mi = np.mean(s)
sigma = np.var(s)
ecdf = ECDF(s)
F = ss.norm.cdf(ecdf.x, loc=mi, scale=sigma)

fig, axes = plt.subplots(1, 2, figsize=(12, 5))

axes[0].plot(np.log(ecdf.x), np.log(1-ecdf.y), label='Dystrybuanta empiryczna', color='darkslateblue')
# axes[0].plot(np.log(ecdf.x), np.log(1-F), label='Dystrybuanta rozkładu normalnego', color='slateblue')
axes[0].set_title('Dane o rozkładzie dwu-modalnym')
axes[0].set_xlabel('log(x)')
axes[0].set_ylabel('log(1-F(x))')
axes[0].legend()

s = PD_eyesOpen[n2]
s = s[(s > 0)]
mi = np.mean(s)
sigma = np.var(s)
ecdf = ECDF(s)
F = ss.norm.cdf(ecdf.x, loc=mi, scale=sigma)

axes[1].plot(np.log(ecdf.x), np.log(1-ecdf.y), label='Dystrybuanta empiryczna', color='darkslateblue')
# axes[1].plot(np.log(ecdf.x), np.log(1-F), label='Dystrybuanta rozkładu normalnego', color='slateblue')
axes[1].set_title('Dane o rozkładzie jedno-modalnym')
axes[1].set_xlabel('log(x)')
axes[1].set_ylabel('log(1-F(x))')
axes[1].legend()

plt.tight_layout()

# Save the figure to a file
fig.savefig("Wykres5a_dystrybuanta_loglog.png")  # Save the entire figure

# Display the plot
plt.show()
```

# DOPASOWANIE ROZKŁADU GEV DO MAKSIMUM

```python
# Create a figure and a 2x2 grid of subplots
fig, axes = plt.subplots(2, 2, figsize=(10, 8))  # 2 rows, 2 columns

window = 100

n1 = 1
n2 = 6

data = PD_eyesOpen[n1]
l = len(data)
maximums1 = np.zeros(len(data)//window+1)

k = 0
while k*window < l:
    d = data[k*window:(k+1)*window]
    maximums1[k] = max(d)
    k += 1
    
shape, loc, scale = gev.fit(maximums1, 0)
x = np.linspace(min(maximums1), max(maximums1), 1000)
ecdf_max = ECDF(maximums1)

# cleaned_HC_ECDF = HC_ECDF.x[~np.isnan(HC_ECDF.x) & ~np.isinf(HC_ECDF.x)]

# Plot data in each subplot
axes[0, 0].plot(ecdf_max.x, ecdf_max.y, label='Dystrybuanta empiryczna', color='darkslateblue')
axes[0, 0].plot(x, gev.cdf(x, shape, loc = loc, scale = scale), color='slateblue', label='Dystrybuanta rozkładu GEV')
axes[0, 0].set_title('Dystrybuanta rozkładu maksimum\n i dopasowany rozkład GEV')
axes[0, 0].set_xlabel('x')
axes[0, 0].set_ylabel('F(x)')
axes[0, 0].legend()

axes[0, 1].hist(maximums1, bins = 'sqrt', density='True', label='Histogram', color='darkslateblue')
axes[0, 1].plot(x, gev.pdf(x, shape, loc = loc, scale = scale), color='slateblue', label='Dystrybuanta rozkładu GEV')
axes[0, 1].set_title("Histogram rozkładu maksimum\n i dopasowany rozkład GEV")
axes[0, 1].set_xlabel('x')
axes[0, 1].set_ylabel('f(x)')
axes[0, 1].legend()

data = HC_eyesOpen[n2]
l = len(data)
maximums2 = np.zeros(len(data)//window+1)

k = 0
while k*window < l:
    d = data[k*window:(k+1)*window]
    maximums2[k] = max(d)
    k += 1
    
shape, loc, scale = gev.fit(maximums2, 0)
x = np.linspace(min(maximums2), max(maximums2), 1000)
ecdf_max = ECDF(maximums2)

axes[1, 0].plot(ecdf_max.x, ecdf_max.y, label='Dystrybuanta empiryczna', color='darkslateblue')
axes[1, 0].plot(x, gev.cdf(x, shape, loc = loc, scale = scale), color='slateblue', label='Dystrybuanta rozkładu GEV')
axes[1, 0].set_title('Dystrybuanta rozkładu maksimum\n i dopasowany rozkład GEV')
axes[1, 0].set_xlabel('x')
axes[1, 0].set_ylabel('F(x)')
axes[1, 0].legend()

axes[1, 1].hist(maximums2, bins = 'sqrt', density='True', label='Histogram', color='darkslateblue')
axes[1, 1].plot(x, gev.pdf(x, shape, loc = loc, scale = scale), color='slateblue', label='Dystrybuanta rozkładu GEV')
axes[1, 1].set_title("Histogram rozkładu maksimum\n i dopasowany rozkład GEV")
axes[1, 1].set_xlabel('x')
axes[1, 1].set_ylabel('f(x)')
axes[1, 1].legend()

# Add labels and other adjustments
# for ax in axes.flat:
#     ax.set_xlabel('Numer obserwacji')
#     ax.set_ylabel('Wartość sygnału')
#     # ax.grid(True)

plt.tight_layout()

# Show the plot
plt.show()

# Optionally, save the figure
fig.savefig("Wykres12_GEV.png")
```

# STATYSTYKA JARQUE-BERA

```python
n1 = 1
n2 = 4

fs = 80000/120
nperseg = 100  # Długość segmentu okna
noverlap = 50  # Liczba próbek nałożenia segmentów

data = PD_eyesOpen[n1]
f, t, Zxx = signal.stft(data, fs=fs, nperseg=nperseg, noverlap=noverlap)

fig, axes = plt.subplots(1, 2, figsize=(12, 5))

K = np.zeros(len(f))
S = np.zeros(len(f))
JB = np.zeros(len(f))
T = len(t)

for j in range(len(f)):
    K[j] = (np.sum((np.abs(Zxx[j, :])**4)))/((np.sum((np.abs(Zxx[j, :])**2)))**2) - 2
    S[j] = (np.sum((np.abs(Zxx[j, :])**3)))/((np.sum((np.abs(Zxx[j, :])**2)))**(3/2))
    JB[j] = T/6*(S[j]**2 + ((K[j]-1)**2)/4)
    
axes[0].plot(f,JB, color='darkslateblue')
# axes[0].colorbar(label='Amplituda')
axes[0].set_title('Statystyka Jarque-Bera - osoba chora')
axes[0].set_ylabel('Statystyka Jarque-Bera')
axes[0].set_xlabel('Częstotliwość')

data = HC_eyesOpen[n2]
f, t, Zxx = signal.stft(data, fs=fs, nperseg=nperseg, noverlap=noverlap)


K = np.zeros(len(f))
S = np.zeros(len(f))
JB = np.zeros(len(f))
T = len(t)

for j in range(len(f)):
    K[j] = (np.sum((np.abs(Zxx[j, :])**4)))/((np.sum((np.abs(Zxx[j, :])**2)))**2) - 2
    S[j] = (np.sum((np.abs(Zxx[j, :])**3)))/((np.sum((np.abs(Zxx[j, :])**2)))**(3/2))
    JB[j] = T/6*(S[j]**2 + ((K[j]-1)**2)/4)
    
axes[1].plot(f,JB, color='darkslateblue')
axes[1].set_title('Statystyka Jarque-Bera - osoba zdrowa')
axes[1].set_ylabel('Statystyka Jarque-Bera')
axes[1].set_xlabel('Częstotliwość')

plt.tight_layout()

# Save the figure to a file
fig.savefig("Wykres10_JB.png")  # Save the entire figure

# Display the plot
plt.show()
```

# WARIANCJA WARUNKOWA

```python
n1 = 1
n2 = 4

fs = 80000/120
nperseg = 100  # Długość segmentu okna
noverlap = 50  # Liczba próbek nałożenia segmentów

data = PD_eyesOpen[n1]
f, t, Zxx = signal.stft(data, fs=fs, nperseg=nperseg, noverlap=noverlap)

fig, axes = plt.subplots(1, 2, figsize=(12, 5))

CVS = np.zeros(len(f))
T = len(t)

for j in range(len(f)):
    s = np.abs(Zxx[j, :])
    N = len(s)
    sigma2 = np.var(s)
    # Wyznaczenie kwantyli dla przedziałów
    x_quantiles = np.quantile(s, [0.004, 0.062, 0.308, 0.692, 0.938, 0.996])
    
    # Przedziały
    A_1 = s[s <= x_quantiles[0]]
    A_2 = s[(s > x_quantiles[0]) & (s <= x_quantiles[1])]
    A_3 = s[(s > x_quantiles[1]) & (s <= x_quantiles[2])]
    A_4 = s[(s > x_quantiles[2]) & (s <= x_quantiles[3])]
    A_5 = s[(s > x_quantiles[3]) & (s <= x_quantiles[4])]
    A_6 = s[(s > x_quantiles[4]) & (s <= x_quantiles[5])]
    A_7 = s[s > x_quantiles[5]]
    sigmas = np.array([np.var(A_1),np.var(A_2),np.var(A_3),np.var(A_4),np.var(A_5),np.var(A_6),np.var(A_7)])
    sigmas = np.nan_to_num(sigmas, nan=0)
    CVS[j] = np.sqrt(N)*((sigmas[2]-sigmas[3]) + (sigmas[4]-sigmas[3]))**2/sigma2

axes[0].plot(f,CVS, color='darkslateblue')
# axes[0].colorbar(label='Amplituda')
axes[0].set_title('Wariancja warunkowa - osoba chora')
axes[0].set_ylabel('Wariancja warunkowa')
axes[0].set_xlabel('Częstotliwość')

data = HC_eyesOpen[n2]
f, t, Zxx = signal.stft(data, fs=fs, nperseg=nperseg, noverlap=noverlap)

CVS = np.zeros(len(f))
T = len(t)

for j in range(len(f)):
    s = np.abs(Zxx[j, :])
    N = len(s)
    sigma2 = np.var(s)
    # Wyznaczenie kwantyli dla przedziałów
    x_quantiles = np.quantile(s, [0.004, 0.062, 0.308, 0.692, 0.938, 0.996])
    
    # Przedziały
    A_1 = s[s <= x_quantiles[0]]
    A_2 = s[(s > x_quantiles[0]) & (s <= x_quantiles[1])]
    A_3 = s[(s > x_quantiles[1]) & (s <= x_quantiles[2])]
    A_4 = s[(s > x_quantiles[2]) & (s <= x_quantiles[3])]
    A_5 = s[(s > x_quantiles[3]) & (s <= x_quantiles[4])]
    A_6 = s[(s > x_quantiles[4]) & (s <= x_quantiles[5])]
    A_7 = s[s > x_quantiles[5]]
    sigmas = np.array([np.var(A_1),np.var(A_2),np.var(A_3),np.var(A_4),np.var(A_5),np.var(A_6),np.var(A_7)])
    sigmas = np.nan_to_num(sigmas, nan=0)
    CVS[j] = np.sqrt(N)*((sigmas[2]-sigmas[3]) + (sigmas[4]-sigmas[3]))**2/sigma2

axes[1].plot(f,CVS, color='darkslateblue')
axes[1].set_title('Wariancja warunkowa - osoba zdrowa')
axes[1].set_ylabel('Wariancja warunkowa')
axes[1].set_xlabel('Częstotliwość')

plt.tight_layout()

# Save the figure to a file
fig.savefig("Wykres11_CVS.png")  # Save the entire figure

# Display the plot
plt.show()
```

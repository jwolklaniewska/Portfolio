# USUWANIE TRENDU I SEZONOWOŚCI Z DANYCH SUROWYCH

```python
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import STL

# Generujemy sztuczny szereg czasowy
np.random.seed(123)
t = np.arange(1, 100)

# Użycie STL do dekompozycji szeregu czasowego
stl = STL(data, seasonal=9, period=8)
result = stl.fit()

# Wykres wyników
fig, (ax1, ax2, ax3, ax4) = plt.subplots(4, 1, figsize=(8, 6))
ax1.plot(result.observed)
ax1.set_title('Dane obserwowane')
ax2.plot(result.trend)
ax2.set_title('Trend')
ax3.plot(result.seasonal)
ax3.set_title('Sezonowość')
ax4.plot(result.resid)
ax4.set_title('Reszty')
plt.tight_layout()
plt.show()

h_max = 100

plot_acf(result.resid, lags=h_max)
plt.xlabel('Opóźnienie')
plt.ylabel('Autokorelacja')
plt.title('ACF')
# plt.ylim(-0.8, 1.1)
# plt.savefig("ACF_5")
plt.show()

plot_pacf(result.resid, lags=h_max)#, method='ywm')
plt.xlabel('Opóźnienie')
plt.ylabel('Częściowa autokorelacja')
plt.title('PACF')
# plt.ylim(-0.8, 1.1)
# plt.savefig("PACF_5")
plt.show()
```

# DEKOMPOZYCJA DANYCH SUROWYCH PRZY UŻYCIU TRANSFORMATY FALKOWEJ

```python
import pywt

coeffs = pywt.wavedec(data, 'db4', level=3)

# Wyciąganie niskich częstotliwości (szum, trend)
trend = coeffs[0]

# Zmiana współczynników dla wyższych częstotliwości (np. ustawienie na zero)
modified_coeffs = [trend] + [np.zeros_like(c) for c in coeffs[1:]]

# Rekonstrukcja sygnału z usuniętym trendem
reconstructed_data = pywt.waverec(modified_coeffs, 'db4')

# Upewnienie się, że długość zrekonstruowanego sygnału jest taka sama jak dane wejściowe
reconstructed_data = reconstructed_data[:len(data)]

plt.plot(data[10000:11000], alpha=0.5)

plt.plot(reconstructed_data[10000:11000])
plt.show()

h_max = 100

plot_acf(reconstructed_data, lags=h_max)
plt.xlabel('Opóźnienie')
plt.ylabel('Autokorelacja')
plt.title('ACF')
# plt.ylim(-0.8, 1.1)
# plt.savefig("ACF_5")
plt.show()

plot_pacf(reconstructed_data, lags=h_max)#, method='ywm')
plt.xlabel('Opóźnienie')
plt.ylabel('Częściowa autokorelacja')
plt.title('PACF')
# plt.ylim(-0.8, 1.1)
# plt.savefig("PACF_5")
plt.show()
```
